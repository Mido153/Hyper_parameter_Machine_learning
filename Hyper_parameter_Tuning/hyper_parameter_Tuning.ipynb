```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
import pandas as pd
data = pd.DataFrame(data=iris.data , columns=iris.feature_names)
data.head()
```

```python
data['flower'] = iris.target
data.head()

# here we need to put the flower name instead of number 
data['flower'] = data['flower'].apply(lambda x: iris.target_names[x])
data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(iris.data , iris.target , test_size = 0.25 )
```

# SVM Model 

```python
# here the hyper parameter tuning is the way to choose the best parameter for the model 
```

```python
from sklearn.svm import SVC 
model = SVC(kernel='rbf' , C=30 , gamma='auto')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
# here we make a different attribute we can calculate the average of the score to know what is the best 
# parameter for our model 
```

```python
import numpy as np
```

```python
# here we make a loop  for simplify 
kernal_val = ['rbf' , 'linear' , 'poly']     # you can try all values 
C_val = [10,20,30]
average_score = {}    # this dictationary for store the avg_score for each Kernal Type 

for kv in kernal_val:
    for cv in C_val:
        score = cross_val_score(SVC(kernel=kv , C= cv , gamma='auto') , iris.data , iris.target , cv=5)
        average_score[kv + '_' + str(cv)] =  np.average(score)    # calculate the average score for each kernal value and store it 
```

```python
# this is the values 
average_score
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
# craete the gridSearchCv Object
clf = GridSearchCV( SVC(gamma='auto'), {
    'C':[10,20,30,40],
    'kernel':['rbf','linear','poly']
}, cv=5 , return_train_score = False )

# now train the model 
clf.fit(iris.data , iris.target)
clf.cv_results_
```

```python
# to easly read the above results do this
df = pd.DataFrame(clf.cv_results_)
df

# some notes 
# splits from 0 to 4 is the K fold number we provided
# rank_test_score is show you what is the best parameter 
# mean_test_score is the average score for this parameter in each cross validation 
```

```python
# simplify the above table 
df[['param_C' , 'param_kernel' , 'mean_test_score']]
```

```python
# the randomizesearch cv is used for make a random iteration
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV( SVC(gamma='auto'), {
    'C':[10,20,30,40],
    'kernel':['rbf','linear','poly']
}, cv=5 , return_train_score = False  , n_iter=2)

# now train the model 
rs.fit(iris.data , iris.target)
df = pd.DataFrame(rs.cv_results_)
df[['param_C' , 'param_kernel' , 'mean_test_score']]
```
